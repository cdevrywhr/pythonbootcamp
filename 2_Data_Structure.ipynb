# **Dictionary ** 

**syntax : {key:value} OR dict({key:value})**

```python
dict_1 = {"Name":"John", "Age":22, "Height":173}
```

```python
type(dict_1)
```

```python
print(dict_1)
```

```python
#sequance of tuple
dict_2 = dict({
          ("John", "22"),
          ("Iris", "22"),
          ("Paul", "22"),
           })
```

```python
print(dict_2)
```

```python
dict_2["John"]
```

```python
dict_2["John"] = "24"
```

```python
print(dict_2)
```
